### Teste detokenisator
"list_tokens.txt" inneholder tokenene som er resultatet av å tokenisere det lille testsettet i grena marie_tokenizer, spesifikt fila "tokenisert_testsett_orig.txt".

```python
with open('list_tokens.txt', 'r') as file:
    tokens = [line.strip() for line in file.readlines()]
```

```python
def detokenize(tokens: list) -> str:
    """Takes a list of tokens and returns a detokenized version."""
    return "".join([
        token if token in """()!?.,:;[]"'«»""" else f" {token}"
        for token in tokens
    ]).strip()
```

```python
detokenized_text = detokenize(tokens)
```

```python
with open('detokenized_text.txt', 'w') as file:
    file.write(detokenized_text)
```
