```python
import os
import pandas as pd
import shutil
```

#### Lese csv-er og skrive til txt-filer

```python
csv_folder_path = '../../maalfrid_eval/corpus'

csv_files = [f for f in os.listdir(csv_folder_path) if f.endswith('.csv')]
csv_files
```

```python
for filename in csv_files:
    csv_path = os.path.join(csv_folder_path, filename)
    df = pd.read_csv(csv_path, header=None)
    df_text = df[2]

    txt_filename = filename.replace('.csv', '.txt')
    txt_path = os.path.join('../../tekster/input/maalfrid', txt_filename)

    with open(txt_path, 'w') as file:
        [file.write(line+'\n') for line in df_text]
```

#### Lese jsonl-filer og skrive til txt-filer

```python
jsonl_folder_path = '../../tekster/stortinget/jsonl'

jsonl_files = [f for f in os.listdir(jsonl_folder_path) if f.endswith('.jsonl')]
jsonl_files
```

```python
for filename in jsonl_files:
    jsonl_path = os.path.join(jsonl_folder_path, filename)
    df = pd.read_json(jsonl_path, lines=True)
    df_text = df['para_text']

    txt_filename = filename.replace('.jsonl', '.txt')
    txt_path = os.path.join('../../tekster/input/stortinget', txt_filename)

    with open(txt_path, 'w') as file:
        [file.write(line+'\n') for line in df_text]
```

#### Lese xml-filer og skrive til txt-filer
Jeg skriver alt innholdet i xml-filen til txt for å få med url-er og andre uvanlige konstruksjoner

```python
xml_folder_path = '../../tekster/bokselskap/202201' #201906  #202101  #202201

xml_files = [f for f in os.listdir(xml_folder_path) if f.endswith('.xml')]
xml_files
```

```python
for filename in xml_files:
    xml_path = os.path.join(xml_folder_path, filename)

    with open(xml_path, 'r') as f:
        text = f.read()

    txt_filename = filename.replace('.xml', '.txt')
    txt_path = os.path.join('../../tekster/input/bokselskap', txt_filename)

    with open(txt_path, 'w') as f:
        f.write(text)
```

#### Hente ut txt-filer fra hver undermappe av nbdigital

```python
source_folder = '../../tekster/nbdigital/idetfrikorpus-dist-txt'
destination_folder = '../../tekster/input/nbdigital'
```

```python
for root, dirs, files in os.walk(source_folder):
    for file in files:
        if file.endswith('.txt'):
            source_path = os.path.join(root, file)
            destination_path = os.path.join(destination_folder, file)

            if os.path.exists(destination_path):
                folder_name = os.path.basename(root)
                new_filename = f"{folder_name}_{file}"
                destination_path = os.path.join(destination_folder, new_filename)
        
            shutil.copy2(source_path, destination_path)
```
