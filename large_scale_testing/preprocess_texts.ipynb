#### Lese xml-filer og skrive til txt-filer
For xml-filene i undermappene /201906 og /202101 henter jeg ut tekstelementer fra xml-en og skriver til txt-fil.\
For xml-filene i undermappa /202201 skriver jeg hele xml-strukturen til txt-fil for å få med uvanlige konstruksjoner.

```python
xml_folder_path = 'texts/corpora/bokselskap/202201' #201906  #202101  #202201

xml_files = [f for f in os.listdir(xml_folder_path) if f.endswith('.xml')]
xml_files
```

```python
def extract_text_from_xml(xml_path):
    """extracts text elements from xml with etree"""

    with open(xml_path, 'rb') as f:
        tree = etree.parse(f)

    root = tree.getroot()
    
    text_elements = []
    for elem in root.iter():
        if elem.text and elem.text.strip():
            text_elements.append(elem.text.strip())
    
    return '\n'.join(text_elements)
```

```python
# kjør extract_text_from_xml på filene i mappene 201906 og 202101

for filename in xml_files:
    xml_path = os.path.join(xml_folder_path, filename)

    text = extract_text_from_xml(xml_path)

    txt_filename = filename.replace('.xml', '.txt')
    txt_path = os.path.join('texts/input/bokselskap', txt_filename)

    with open(txt_path, 'w', encoding='utf-8') as f:
        f.write(text)
```

```python
# konverter xml til txt for filene i mappa 202201

for filename in xml_files:
    xml_path = os.path.join(xml_folder_path, filename)

    with open(xml_path, 'r') as f:
        text = f.read()

    txt_filename = filename.replace('.xml', '.txt')
    txt_path = os.path.join('texts/input/bokselskap', txt_filename)

    with open(txt_path, 'w') as f:
        f.write(text)
```

#### Hente ut txt-filer fra hver undermappe av nbdigital

```python
source_folder = 'texts/corpora/nbdigital/idetfrikorpus-dist-txt'
destination_folder = 'texts/input/nbdigital'
```

```python
for root, dirs, files in os.walk(source_folder):
    for file in files:
        if file.endswith('.txt'):
            source_path = os.path.join(root, file)
            destination_path = os.path.join(destination_folder, file)

            if os.path.exists(destination_path):
                folder_name = os.path.basename(root)
                new_filename = f"{folder_name}_{file}"
                destination_path = os.path.join(destination_folder, new_filename)
        
            shutil.copy2(source_path, destination_path)
```
