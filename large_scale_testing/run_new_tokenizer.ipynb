## Tokenisere med ny tokenisator

```python
import sys
import time
import os
```

```python
parent_dir = os.path.abspath(os.path.join(os.getcwd(), "../"))
sys.path.append(parent_dir)

from src import nb_tokenizer as nb
```

```python
# folder = 'bokselskap'  'maalfrid_eval'  'nbdigital'  'stortinget'

folder = 'bokselskap'

input_folder = 'texts/input/' + folder
output_folder = 'texts/output/ny_tokenisator/' + folder
```

```python
input_files = [f for f in os.listdir(input_folder) if f.endswith('.txt')]
```

```python
t0 = time.process_time()

for filename in input_files:
    input_path = os.path.join(input_folder, filename)

    with open(input_path, 'r') as file:
        text = file.read()

    tokens = nb.tokenize(text)

    output_filename = filename.replace('.txt', '_ny.txt')
    output_path = os.path.join(output_folder, output_filename)

    with open(output_path, 'w') as file:
        [file.write(i+'\n') for i in tokens]

t1 = time.process_time()
t = t1 - t0

with open('tokenizer_times.txt', 'a') as file:
    file.write(f'\ntid ny {folder}: {t}')
```
